# find & findall function work in REGEX

```python
#demo of findall()method
#extract email ID from a given string
import re
s='A msg from admin@footprint.co to nagendra.yadav@footprint.co abt meeting at 14:00'
lst=re.findall('\S+@\S+',s)
print(lst)
```

```python
import re
fhand= open('D:/Local Disk/Python Code/BAISALI/mbox-short.txt')
ls1=list()
for line in fhand:
    line=line.strip()
    lst=re.findall('\S+@\S+',line)
    if(len(lst))>0:
        ls1.extend(lst)
for i in range(len(ls1)):
    ls1[i]=ls1[i].strip('<>;)')
print(ls1)
fhand.close()
    
```
